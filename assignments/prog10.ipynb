## 2. El Nio Southern Oscillation

The El Nio Southern Oscillation (ENSO) is the largest mode of interannual variability in Earth's ocean and atmosphere. During El Nio conditions, water in eastern equatorial Pacific Ocean, near the coast of South America, is warmer than usual and atmospheric deep convection shifts eastward over the Pacific Ocean. La Nia conditions bring cold water to the eastern Pacific and convection shifts westward. Both El Nio and La Nia affect weather patterns, espeically precipitation, around the globe.
